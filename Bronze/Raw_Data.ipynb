```python
claims_df = spark.read.option("header", True).csv(base_path + "claims.csv")
customers_df = spark.read.option("header", True).csv(base_path + "customrs.csv")
fnol_df = spark.read.option("header", True).csv(base_path + "fnol.csv")
payments_df = spark.read.option("header", True).csv(base_path + "paymnts.csv")
policies_df = spark.read.option("header", True).csv(base_path + "policies.csv")
vehicles_df = spark.read.option("header", True).csv(base_path + "vehiclees.csv")
```

##Save As Tables in Vehicle_Insurance Catalog

```python
claims_df.write.format("delta").mode("overwrite").saveAsTable("vehicle_insurance.bronze.claims")

customers_df.write.format("delta").mode("overwrite").saveAsTable("vehicle_insurance.bronze.customers")

fnol_df.write.format("delta").mode("overwrite").saveAsTable("vehicle_insurance.bronze.fnol")

payments_df.write.format("delta").mode("overwrite").saveAsTable("vehicle_insurance.bronze.payments")

policies_df.write.format("delta").mode("overwrite").saveAsTable("vehicle_insurance.bronze.policies")

vehicles_df.write.format("delta").mode("overwrite").saveAsTable("vehicle_insurance.bronze.vehicles")
```
